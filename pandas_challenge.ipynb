```python
import pandas as pd
import numpy as np
```

```python
#Cargar el archivo anime.csv como dataframe
df = pd.read_csv('anime.csv')
df.head(50)
```

```python
df.dtypes
```

```python
df.isnull()
```

```python
#Reemplazar los valores nulos segun su tipo:
#los numericos por el promedio de la Columna y los strings por 'none'

df["anime_id"].fillna(df["anime_id"].mean(), inplace = True) 
df['anime_id'].isnull().values.any()

df["c1"].fillna(df["c1"].mean(), inplace = True) 
df['c1'].isnull().values.any()

df["c2"].fillna(df["c2"].mean(), inplace = True) 
df['c2'].isnull().values.any()

df["episodes"].fillna(df["episodes"].mean(), inplace = True) 
df['episodes'].isnull().values.any()

df["members"].fillna(df["members"].mean(), inplace = True) 
df['members'].isnull().values.any()

df["rating"].fillna(df["rating"].mean(), inplace = True) 
df['rating'].isnull().values.any()
```

```python
#Reemplazar los valores nulos segun su tipo:
#los numericos por el promedio de la Columna y los strings por 'none'
def replace_null_str(text):
  if text is np.nan:
    return 'none'
  return text

filtered_columns = df.dtypes[df.dtypes == np.object]
obj_columns = list(filtered_columns.index)
obj_columns
for i in obj_columns:
    df[i] = df[i].apply(replace_null_str)
```

```python
df.isnull().values.any()
```

```python
#Obtener suma total de 'members'
df["members"].sum()
```

```python
cols = ['type','source']
df[cols] = df[cols].astype('category')
```

```python
#Obtener el rating promedio de cada tipo de serie
types=['Movie', 'Music', 'ONA', 'OVA', 'Special', 'TV', 'none']
for i in types:
    type_music = df[df["type"].isin([i])]
    rating_mean=type_music['rating'].mean()
    print('Type: ',i)
    print('mean rating: ',rating_mean)
```

```python
#Obtener el maximo rating de series con type 'Music'
music_rating = df[df["type"].isin(['Music'])]
rating_max=music_rating['rating'].max()
print('El maximo rating de series con tipo music es: ',rating_max)
```

```python
#(Bonus)Convierta las columnas genre, main_cast y studios en listas 
#ej. "Drama,Romance,School,Supernatural" -> [Drama, Romance, School, Supernatural] 
df["genre"]= df["genre"].str.split(",")
df["main_cast"]= df["main_cast"].str.split(";")
df["studios"]= df["studios"].str.split(",")

df.head()
```
